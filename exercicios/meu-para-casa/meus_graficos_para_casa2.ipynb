```python
def processamento_df(df_amazon):

    df_amazon = df_amazon.drop([ "review_id", "review_title", "review_content", "img_link",	"product_link","about_product","rating_count"], axis = "columns")
    df_amazon = df_amazon.sort_values("discount_percentage")
    df_amazon = df_amazon.reset_index()

    metricas= df_amazon.describe()

    return df_amazon,metricas
```

```python
df_amazon_sem_Na,metricas  = processamento_df(df_amazon)
```

```python
df_amazon_sem_Na
```

```python
media_rating = df_amazon_sem_Na['rating'].mean()
```

```python
media_grafico= round(media_rating,2)
```

```python
contagem_avaliações = df_amazon_sem_Na['rating'].value_counts()
```

```python
contagem_avaliações.plot(kind='bar',edgecolor = 'black', color = 'green')
plt.axvline(x=media_grafico, color='r', linestyle='--', label=f'Média das Avaliações ({media_grafico})')

plt.title("Número de avaliações")
plt.xlabel("Avaliações")
plt.ylabel("Quantidade")

plt.legend()
```
